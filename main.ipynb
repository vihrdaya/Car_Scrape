```python
def extract_data(section):
    label_items = []
    label = section.find_all(class_='formLabelColumn')
    value = section.find_all(class_='pfFormControl')
    
    for i in range(len(label)):
        label_spans = label[i].find_all(class_='inquiry-label')
        value_spans = value[i].find_all(class_='inquiry-inp')
        
        for element, inp in zip(label_spans, value_spans):
            text_label = element.get_text(separator=' ', strip=True) or 'N/A'
            text_value = inp.get_text(separator=' ', strip=True) or 'N/A'
            label_items.append((text_label, text_value))
            
    return label_items

def process_sections(soup, section_ids):
    results = []
    for section_id in section_ids:
        section = soup.find(id=section_id)
        if section:
            results.extend(extract_data(section))

    return results
```
